```python
%pylab inline
import seaborn as sns
sns.set_palette("magma",8)
```

```python
dxi=0.0001
N=800000
for n in range(4,20):
    xi=0.01
    theta=1.0
    f1=0.0
    theta_sol=[]
    xi_sol=[]
    for i in range(N):
        f1+=-xi**2*theta**(n)*dxi
        theta+=f1/xi**2*dxi
        xi+=dxi
        theta_sol.append(theta)
        xi_sol.append(xi)
        if(theta_sol[i]<0 and theta_sol[i-1]>=0):
            print(n,xi)
        if(theta_sol[i]>0 and theta_sol[i-1]<=0):
            print(n,xi)    
    plot(xi_sol,theta_sol)
xlim(0,15)
ylim(-0.1,1.)
axhline(y=0.0,c="black")
show()
```
